```python
import pandas as pd
import csv
# Three CSV files
March = pd.read_csv("March LA -2024(new).csv")
April = pd.read_csv("April LA 2024.csv")
May = pd.read_csv("May LA 2024.csv")
#Merging the three into one data frame
LA = pd.concat([March,April,May],ignore_index=True)
LA
```

```python
LA.info()
#Checking whether there are null values
LA.shape
```

```python
#Dropping columns with null values
LA.drop(columns=['country_iso', 'isMobile', 'isMonitor'], inplace=True)
```

```python

# Converting the 'datetimeUtc' column to datetime format
LA['datetimeUtc'] = pd.to_datetime(LA['datetimeUtc'])

# To check the format of the 'datetimeUtc'
LA.info()
```

```python
# Extracting Dates from datetimeUtc
LA['date_only'] = LA['datetimeUtc'].dt.date

# Extracting hour from datetimeUtc
LA['hour_only'] = LA['datetimeUtc'].dt.hour
```

```python


# Group by 'parameter' 
# Within each parameter groupby date
summary_stat = LA.groupby(['parameter', 'datetimeUtc'])['value'].agg(['mean', 'min', 'median', 'max']).reset_index()

# To print the first 10 results
print(summary_stat.head(21))
```

```python
#VISUALIZATIONS
LA['date_only'] = LA['datetimeUtc'].dt.date
LA['hour_only'] = LA['datetimeUtc'].dt.hour
LA['month_only'] = LA['datetimeUtc'].dt.month
```

```python
import matplotlib.pyplot as plt
import pandas as pd

nitrogen_dioxide = LA[LA['parameter']=='no2']
ozone = LA[LA['parameter']=='o3']

daily_aqi = nitrogen_dioxide.groupby(['month_only', 'date_only'])['value'].mean()
daily_aqi_1 = ozone.groupby(['month_only', 'date_only'])['value'].mean()

plt.figure(figsize=(15, 7)) 
daily_aqi.plot(kind='bar', color='red')
plt.title('Average AQI of March, April, May')
plt.xlabel('Day')
plt.ylabel('Average AQI')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(nitrogen_dioxide['value'], bins=30, edgecolor='k')
plt.title('Histogram of NO2 Levels in Los Angeles')
plt.xlabel('NO2 Concentration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Time series plot
plt.figure(figsize=(12, 6))
plt.plot(nitrogen_dioxide.index, nitrogen_dioxide['value'], label='NO2 levels')
plt.title('Time Series of NO2 Levels in Los Angeles')
plt.xlabel('Date')
plt.ylabel('NO2 Concentration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.subplot(4, 1, 1)
plt.hist(nitrogen_dioxide['value'], bins=30, edgecolor='k')
plt.title('Histogram of NO2 Levels in Los Angeles')
plt.xlabel('NO2 Concentration')
plt.ylabel('Frequency')
plt.grid(True)
'''The histogram reveals how NO2 concentrations are distributed. 
Peaks in the histogram indicate the most common NO2 levels, 
while the spread of the histogram shows the range of NO2 concentrations in Los Angeles.'''
```

```python
import seaborn as sns
pivot= pd.pivot_table(LA, 
                      index=['datetimeUtc', 'location_name', 'unit'], 
                      columns='parameter', 
                      values='value', 
                      aggfunc='mean')

pivot.reset_index(inplace=True)
pivot = pivot[['location_name', 'datetimeUtc', 'o3','no2','unit']]

# print(pivot.head(5))

pollutants = ['no2', 'o3']
correlation_matrix = pivot[pollutants].corr()

plt.figure(figsize=(10, 8))
plt.title('Correlation Matrix of Pollutants')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.show()

'''The heatmap provides a comprehensive view of how NO2 levels have changed over time. 
Each cell represents the average NO2 concentration for a specific month and year. 
Darker colors (depending on the color map used) might indicate higher NO2 levels. 
This visualization helps identify any long-term trends or anomalies in NO2 pollution.
'''
```
