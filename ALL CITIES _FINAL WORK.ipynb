#Calculate AQI fro each pollutant and add them as new columns

#Los_Angeles

```python
# Define breakpoints for CO (Carbon Monoxide) in µg/m³
# Each tuple represents (low concentration, high concentration, low AQI, high AQI)
breakpoints_co = [(0.0, 5000.0, 0, 50), 
                  (5001.0, 10000.0, 51, 100), 
                  (10001.0, 13000.0, 101, 150), 
                  (13001.0, 15500.0, 151, 200), 
                  (15501.0, 30400.0, 201, 300), 
                  (30401.0, 40400.0, 301, 400), 
                  (40401.0, 50400.0, 401, 500)]

# Define breakpoints for O3 (Ozone), PM2.5 (Particulate Matter 2.5), and NO2 (Nitrogen Dioxide)
# Convert ppm to µg/m³ using the conversion factor
ppm_to_ugm3 = 2140
breakpoints_O3 = [(0.000 * ppm_to_ugm3, 0.054 * ppm_to_ugm3, 0, 50),
                  (0.055 * ppm_to_ugm3, 0.070 * ppm_to_ugm3, 51, 100),
                  (0.071 * ppm_to_ugm3, 0.085 * ppm_to_ugm3, 101, 150),
                  (0.086 * ppm_to_ugm3, 0.105 * ppm_to_ugm3, 151, 200),
                  (0.106 * ppm_to_ugm3, 0.200 * ppm_to_ugm3, 201, 300),
                  (0.201 * ppm_to_ugm3, 0.404 * ppm_to_ugm3, 301, 400),
                  (0.405 * ppm_to_ugm3, 0.504 * ppm_to_ugm3, 401, 500)]

# Breakpoints for PM2.5 in µg/m³
breakpoints_pm25 = [(0, 12.0, 0, 50), 
                    (12.1, 35.4, 51, 100), 
                    (35.5, 55.4, 101, 150),
                    (55.5, 150.4, 151, 200), 
                    (150.5, 250.4, 201, 300), 
                    (250.5, 350.4, 301, 400),
                    (350.5, 500.4, 401, 500)]

# Breakpoints for NO2 in µg/m³
breakpoints_No2 = [(0, 53, 0, 50), 
                   (54, 100, 51, 100), 
                   (101, 360, 101, 150),
                   (361, 649, 151, 200), 
                   (650, 1249, 201, 300), 
                   (1250, 1649, 301, 400),
                   (1650, 2049, 401, 500)]

# Function to calculate AQI (Air Quality Index) based on concentration and breakpoints
def calculate_aqi(concentration, breakpoints):
    for bp in breakpoints:
        if bp[0] <= concentration <= bp[1]:  # Check if concentration falls within the breakpoint range
            C = concentration
            BP_hi = bp[1]  # High breakpoint concentration
            BP_lo = bp[0]  # Low breakpoint concentration
            I_hi = bp[3]  # High AQI value
            I_lo = bp[2]  # Low AQI value
            # Calculate AQI using the linear interpolation formula
            aqi = ((I_hi - I_lo) / (BP_hi - BP_lo)) * (C - BP_lo) + I_lo
            return round(aqi)  # Return the rounded AQI value
    return None  # Return None if concentration is outside defined breakpoints

# Calculate AQI for each pollutant in the Los_Angeles_pivot DataFrame
Los_Angeles_pivot['AQI_O3'] = Los_Angeles_pivot['O3'].apply(lambda x: calculate_aqi(x, breakpoints_O3))
Los_Angeles_pivot['AQI_PM25'] = Los_Angeles_pivot['pm25'].apply(lambda x: calculate_aqi(x, breakpoints_pm25))
Los_Angeles_pivot['AQI_NO2'] = Los_Angeles_pivot['No2'].apply(lambda x: calculate_aqi(x, breakpoints_No2))
Los_Angeles_pivot['AQI_CO'] = Los_Angeles_pivot['Co'].apply(lambda x: calculate_aqi(x, breakpoints_co))

# Calculate the overall AQI by taking the maximum AQI value from each pollutant
Los_Angeles_pivot['AQI'] = Los_Angeles_pivot[['AQI_O3', 'AQI_PM25', 'AQI_NO2', 'AQI_CO']].max(axis=1)

# Save the updated DataFrame with AQI calculations to a new CSV file
Los_Angeles_pivot.to_csv('pivot_Los_Angeles.csv', index=False)

# Print the first few rows of the updated DataFrame to verify the changes
print(Los_Angeles_pivot.head())
```

# Heatmap for AQI levels over the three months (March, April, and May) for the eight cities

Best Air Quality: Los Angeles, Barcelona, Toronto, and Sofienbergparken, with consistently good air quality.

Poor Air Quality: Beijing and Cairo, consistently unhealthy for sensitive groups.

Deteriorating Air Quality: Nairobi and Baghdad, showing a significant increase in AQI levels and worsening air quality over time.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Function to aggregate AQI data by month
def aggregate_aqi_by_month(df):
    df['date'] = pd.to_datetime(df['datetimeUtc'])  # Convert 'datetimeUtc' column to datetime format
    df['month'] = df['date'].dt.month  # Extract the month from the 'date' column
    monthly_avg_aqi = df.groupby('month')['AQI'].mean()  # Group by month and calculate the mean AQI
    return monthly_avg_aqi

# Aggregate AQI data for each city
la_monthly_aqi = aggregate_aqi_by_month(Los_Angeles_pivot)
barcelona_monthly_aqi = aggregate_aqi_by_month(Barcelona_pivot)
beijing_monthly_aqi = aggregate_aqi_by_month(Beijing_pivot)
nairobi_monthly_aqi = aggregate_aqi_by_month(Nairobi_pivot)
toronto_monthly_aqi = aggregate_aqi_by_month(Toronto_pivot)
cairo_monthly_aqi = aggregate_aqi_by_month(Cairo_pivot)
baghdad_monthly_aqi = aggregate_aqi_by_month(Baghdad_pivot)
sofienbergparken_monthly_aqi = aggregate_aqi_by_month(Sofienbergparken_pivot)

# Create a DataFrame for the heatmap
aqi_data = pd.DataFrame({
    'Los Angeles': la_monthly_aqi,
    'Barcelona': barcelona_monthly_aqi,
    'Beijing': beijing_monthly_aqi,
    'Nairobi': nairobi_monthly_aqi,
    'Toronto': toronto_monthly_aqi,
    'Cairo': cairo_monthly_aqi,
    'Baghdad': baghdad_monthly_aqi,
    'Sofienbergparken': sofienbergparken_monthly_aqi
}).transpose()

# Rename columns to represent months
aqi_data.columns = ['March', 'April', 'May']

# Plotting the heatmap with formatted annotations
plt.figure(figsize=(16, 6))  # Create a new figure with a specified size (16 inches wide, 6 inches tall)
sns.heatmap(aqi_data, annot=True, fmt=".1f", cmap='coolwarm', linewidths=.5, annot_kws={"size": 10})  
# Create a heatmap with formatted annotations, a coolwarm color map, and specific annotation size
plt.title('Monthly Average AQI Levels for Various Cities (March - May 2024)')  # Set the title of the heatmap
plt.xlabel('Month')  # Label the x-axis as 'Month'
plt.ylabel('City')  # Label the y-axis as 'City'
plt.show()  # Display the heatmap
```
