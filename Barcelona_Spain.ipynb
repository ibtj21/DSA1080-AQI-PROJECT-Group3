## Barcelona(Ciutadella) Spain

The timeframe is March 2024, April 2024, May 2024

The pollutants measured are NO2(Nitrogen Dioxide) and O3(Ozone)

##
Importing the necessary packages
##

```python
import csv
import pandas as pd       # for data manipulation and analysis
import numpy as np        # for numerical computing
import matplotlib.pyplot as plt  # for data visualization
import seaborn as sns     # for data visualization
```

##
Reading the csv files into a Pandas DataFrame
##

```python
# Reading the csv files into a DataFrame
mar = pd.read_csv('2024_Mar_Barcelona_Spain.csv')
apr = pd.read_csv('2024_Apr_Barcelona_Spain.csv')
may = pd.read_csv('2024_May_Barcelona_Spain.csv')

# Merging the different files into a single DataFrame
barcelona = pd.concat([mar,apr,may],ignore_index=True)

# Printing the first five observations
barcelona.head()
```

## Data Description ##

```python
# To find the total number of rows and columns in the data set
barcelona.shape

# It has 3426 rows/observations and 15 columns/characheristics
```

```python
# To know the columns and their corresponding data_types
# To know whether the data contains missing values or not
barcelona.info()
```

```python
'''
It is seen that columns such as 'country_iso', 'isMobile', 'isMonitor' have null values
thus dropping them.
'''
barcelona.drop(columns=['country_iso', 'isMobile', 'isMonitor', 'owner_name', 'provider'], inplace=True)
```

```python
barcelona.info()
barcelona.head()
```

```python
'''
As seen from above, datetimeUtc data type is object/string.
It is therefore necessary to convert the datatype to datetime format
so as to perform time-series based operation on the dates.
'''
# Converting the 'datetimeUtc' column to datetime format
barcelona['datetimeUtc'] = pd.to_datetime(barcelona['datetimeUtc'])

# To check the format of the 'datetimeUtc'
barcelona.info()
```

```python
# To check for duplicated rows
barcelona.duplicated().sum()

# There are no duplicated rows
```

```python
# A final check on missing values
barcelona.isnull().sum()

# There are no null values
```

## Summary Statistics ##

```python
# Extracting Dates from datetimeUtc
barcelona['date_only'] = barcelona['datetimeUtc'].dt.date

# Extracting hour from datetimeUtc
barcelona['hour_only'] = barcelona['datetimeUtc'].dt.hour
```

```python
# Getting summary statistics

# Group by 'parameter' first
# Within each parameter groupby date
# To get the 'mean' 'median' 'min' 'max' for each category
summary_stats = barcelona.groupby(['parameter', 'date_only'])['value'].agg(['mean', 'min', 'median', 'max']).reset_index()

# To print the first 10 results
print(summary_stats.head(11))
```

##
Bar Graph
##

```python
# Extracting Dates from datetimeUtc
barcelona['date_only'] = barcelona['datetimeUtc'].dt.date

# Extracting Hour from datetimeUtc
barcelona['hour_only'] = barcelona['datetimeUtc'].dt.hour

# Extracting Month from datetimeUtc
barcelona['month_only'] = barcelona['datetimeUtc'].dt.month
```

```python
# Filtering the DataFrame 'barcelona' to output a subset of rows where
# the parameter is specifically no2
nitrogen_dioxide = barcelona[barcelona['parameter']=='no2']

# Filtering the DataFrame 'barcelona' to output a subset of rows where
# the parameter is specifically o3
ozone = barcelona[barcelona['parameter']=='o3']


# grouping the data by month then by dates and calculating the mean of 'value'
# for each subset of data
daily_aqi = nitrogen_dioxide.groupby(['month_only', 'date_only'])['value'].mean()
daily_aqi_1 = ozone.groupby(['month_only', 'date_only'])['value'].mean()



plt.figure(figsize=(15, 7))  # the size of the output figure
daily_aqi.plot(kind='bar', color='skyblue')  # using data from 'daily_aqi' to plot a bar chart in skyblue color
# Giving the chart a revelant title
plt.title('Average AQI for NO2 by Day for Barcelona for the months of March, April, May')
# Labeling the x-axis
plt.xlabel('Day')
# Labeling the y-axis
plt.ylabel('Average AQI')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# To display the chart
plt.show()
```

```python
plt.figure(figsize=(15, 7)) # the size of the output figure
daily_aqi_1.plot(kind='bar', color='skyblue') # using data from 'daily_aqi_1' to plot a bar chart in skyblue color
# Giving the chart a revelant title
plt.title('Average AQI for O3 by Day for Barcelona for the months of March, April, May')
# Labeling the x-axis
plt.xlabel('Day')
# Labeling the y-axis
plt.ylabel('Average AQI')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# To display the chart
plt.show()
```

##
As seen from the above two bar plots, ozone tends to generally have higher 
emissions than nitrogen dioxide in Barcelona.

##
Time Series analysis
##

```python
# Filtering the DataFrame 'barcelona' to output a subset of rows where
# the parameter is specifically no2
nitrogen_dioxide = barcelona[barcelona['parameter']=='no2']

# Filtering the DataFrame 'barcelona' to output a subset of rows where
# the parameter is specifically o3
ozone = barcelona[barcelona['parameter']=='o3']

# grouping the data by dates and calculating the mean of 'value'
# for each subset of data
daily_aqi = nitrogen_dioxide.groupby(['date_only'])['value'].max()
daily_aqi_1 = ozone.groupby(['date_only'])['value'].max()

# the size of the output figure
plt.figure(figsize=(15, 7))
# using data from 'daily_aqi' to plot a line chart in skyblue color
daily_aqi.plot(kind='line', color='skyblue')
# Giving the chart a revelant title
plt.title('Time Series Analysis for NO2 for Barcelona for the months of March, April, May')
# Labeling the x-axis
plt.xlabel('Day')
# Labeling the y-axis
plt.ylabel('Value')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# To display the chart
plt.show()
```

```python
'''
Over the three month period the AQI for no2 mostly ranges between 80 to 40 
showing the air quality is likely 'fair'
'''
```

```python
# the size of the output figure
plt.figure(figsize=(15, 7))
# using data from 'daily_aqi_1' to plot a line chart in skyblue color
daily_aqi_1.plot(kind='line', color='skyblue')
# Giving the chart a revelant title
plt.title('Time Series Analysis for O3 for Barcelona for the months of March, April, May')
# Labeling the x-axis
plt.xlabel('Day')
# Labeling the y-axis
plt.ylabel('Value')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# To display the chart
plt.show()
```

```python
'''
Over the three month period the AQI for o3 mostly ranges between 80 to 120 
showing the air quality is likely 'poor'
'''
```

##
Heatmaps
##

```python
# Pivot the DataFrame with multi-level index including 'location_name' and 'unit'
pivot= pd.pivot_table(barcelona, 
                      index=['datetimeUtc', 'location_name', 'unit'], 
                      columns='parameter', 
                      values='value', 
                      aggfunc='mean')

# Reset index to convert the multi-level index back into columns
pivot.reset_index(inplace=True)
# Reorder columns 
pivot = pivot[['location_name', 'datetimeUtc', 'o3','no2','unit']]

# print(pivot.head(5))

# Creating a heatmap
pollutants = ['no2', 'o3']
correlation_matrix = pivot[pollutants].corr()

plt.figure(figsize=(10, 8))
plt.title('Correlation Matrix of Pollutants')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.show()
```
